## Problema 11. [Recta tangente]
Lean el siguiente recuadro:

<div style="border:2px solid black; padding:10px; border-radius:10px;">
<h4>Recta tangente:</h4> <p>La derivada de una función $f$ en un punto $x_0$ de su dominio (en el que $f$ es derivable) es la pendiente de la recta tangente al gráfico de $f$ en el punto $\left(x_0, f(x_0)\right)$. La figura muestra el gráfico de una función $f$ y su recta tangente en dos puntos distintos.</p>

|                          |                           |
|--------------------------|---------------------------|
| ![](img/image-028.png)   | ![](img/image-030.png)    |
|                          |                           |

<p>Esa recta tangente tiene una ecuación del tipo $y = mx + b$.</p>
<p>Pero, como su pendiente es $m = f'(x_0)$, la ecuación queda:</p>
<table style="width:100%;">
  <tr>
    <td style="text-align:center;">$y = f'(x_0)x + b$</td>
    <td style="text-align:right; width:80px;">(2.1)</td>
  </tr>
</table>

<p>Además, pasa por el punto $\left( x_0,  f(x_0) \right)$. Poniendo esta condición en $\left(2.1\right)$ se tiene que:</p>

<table style="width:100%;">
    <tr>
        <td style="text-align:center;">$f(x_0) = f'(x_0)x_{0} + b$</td>
    </tr>
</table>
<p>de donde, despejando $b$:</p>
<table style="width:100%;">
    <tr>
        <td style="text-align:center;">$b = f(x_0) - f'(x_0)x_{0}$</td>
        <td style="text-align:right; width:80px;">(2.2)</td>
    </tr>
</table>
<p>Poniendo esto en $\left(2.1\right)$, resulta:</p>
<table style="width:100%;">
    <tr>
        <td style="text-align:center;">$y = f'(x_0)x + f(x_0) - f'(x_0)x_{0}$</td>
        <td style="text-align:right; width:80px;">(2.3)</td>
    </tr>
</table>
<p>y, tomando factor común $f'(x_0)$, y ordenando, obtenemos, finalmente, la ecuación de la recta tangente al gráfico de $f$ en el punto $\left(x_{0},f(x_{0})\right)$:</p>

<table style="width:100%;">
    <tr>
        <td style="text-align:center;"><div style="border:1px solid #ccc; padding:10px; border-radius:5px; display: inline-block;">$y = f'(x_0)(x-x_0) + f(x_0)$</div></td>
        <td style="text-align:right; width:80px">(2.4)</td>
    </tr>
</table>

</div>    

<br>

1. Escriban la ecuación de la recta que es tangente al gráfico de $f$ en el punto $\left(4, 1 \right)$.
